```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

# Distribution Plot

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

# JoinPlots

```python
sns.jointplot(x='total_bill',y='tip',data=tips)#default scatter
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex') #hex
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') #regression
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde') #kde
```

# Pairplot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex') #pass the hue as category
```

# rugplot

```python
sns.rugplot(tips['total_bill'])
```
