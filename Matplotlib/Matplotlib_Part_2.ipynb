```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y= x**2
```

```python
fig =plt.figure()
axes =fig.add_axes([0.1,0.1,0.8,0.8])  ##[Left, buttom, height, width]
```

```python
#axes.plot(x,y)
#for more plots , to remove overlaps 
#plt.tight:layout()
fig,axes =plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)#indexing 
axes[1].plot(y,x)
```

# Figure Size and DPi

```python
fig =plt.figure(figsize=(3,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,'r',label='xy')
ax.plot(y,x,'b',label='yx')
ax.legend(loc=10)#centre #see documentation
#ax.legend(loc=(0.1,0.1)) ##specify as tuples
```

```python
fig,axes =plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

# Costumizing the plots

```python
fig = plt.figure()
ax =fig.add_axes([0,0,1,1])
#ax.plot(x,y,color='blue',linewidth=3,linestyle='--',marker='o')
#aplha =transparent
ax.plot(x,y,color='blue',lw=3,ls='--',alpha = 0.5,marker='o',markersize=20,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')
```

```python
fig = plt.figure()
ax =fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',linewidth=3,linestyle='--',marker='o')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
