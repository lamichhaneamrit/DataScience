```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y= x**2
```

```python
#FuNctional method
plt.plot(x,y)
#plt.show()
plt.xlabel('X Ticks')
plt.ylabel('Y Ticks')
plt.title('My Title')
```

```python
#Multiplots
```

```python
plt.subplot(1,2,1) #(no.of rows, no.of columns , plot number )
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# OOP Methods 

```python
fig =plt.figure()
axes =fig.add_axes([0.1,0.1,0.8,0.8])  ##[Left, buttom, height, width]
axes.plot(x,y)
axes.set_xlabel('X Ticks')
axes.set_ylabel('Y Ticks')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Bigger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```
